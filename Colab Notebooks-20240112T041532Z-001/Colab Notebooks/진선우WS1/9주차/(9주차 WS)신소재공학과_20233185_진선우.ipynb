##3.지도와 Marker 표시하기



###1)지도 생성하기

```python
#지도를 생성하기 위한 folium 라이브러리 선언하기
#import folium
import folium
```

```python
#folium.Map()을 사용하여 지도 생성하고 출력하기
#변수명 = folium.Map(속성들)
    #location=[위도, 경도] : 지도의 중심 좌표를 [위도, 경도] 또는(위도, 경도)로 나타냄
    #zoom_start = 정수 : 지도를 처음 그릴때 확대 정도
    #zoom_control=True : zoom in/out 버튼 표시 여부(True/False)
    #control_scales=False : 스케일 컨트롤 버튼 표시 여부(True/False)
    #tiles=‘스타일’: 지도 스타일을 지정(openstreetmap,stamenterrain,cartodbpositron)

parkingMap = folium.Map(location = [37.4974, 126.9563],  #중심좌표설정(구글 맵 이용)
                        zoom_start = 13,
                        zoom_control = False,
                        control_scale = True)
parkingMap
```

###2)지도에 Marker 표시하기

```python
#DataFrame에서 각 index의 column 데이터를 이용하여 Marker 표시하기
#for index, 행정보 in 변수명.iterrows(): 변수명에 저장된 데이터를 index 단위로 읽어옴 (유료주차장 데이터 사용 예정)
    #변수1 = 행정보['열이름']
    # ....

    #변수명 = folium.Marker(속성들)
        #location=[위도, 경도] : marker가 표시되는 위치정보(위도,경도)
        #popup=‘문자열’: marker를 클릭했을때 나타나는 메세지(문자열)
        #tooltip=‘문자열’: marker에 마우스를 올렸을때 나타나는 메세지(문자열)
        #icon=‘모양’: marker의 모양을 설정(https://fontawesome.com/search?s=thin&o=r)
#변수명.add_to(지도변수명)

for index, 행정보 in data.iterrows():  #유료주차장 데이터 사용 예정
  주차장명 = 행정보['주차장명']
  위도 = 행정보['위도']
  경도 = 행정보['경도']

  mk = folium.Marker(location = [위도, 경도],
                     tooltip = 주차장명,   #속성을 popup 으로 바꿔보기...다른점은?
                     icon =  folium.Icon(color='red', icon ='car', prefix='fa' ))

  mk.add_to(parkingMap)

parkingMap
```

##4.지도에 MarkerCluster와 Marker 표시하기


###1)Marker와 MarkerCluster 를 표시할 지도 생성하기

```python
#Marker와 MarkerCluster 를 표시할 지도 생성하기
#변수명 = folium.Map(속성들)
    #location=[위도, 경도] : 지도의 중심 좌표를 [위도, 경도] 또는(위도, 경도)로 나타냄
    #zoom_start = 정수 : 지도를 처음 그릴때 확대 정도
    #zoom_control=True : zoom in/out 버튼 표시 여부(True/False)
    #control_scales=False : 스케일 컨트롤 버튼 표시 여부(True/False)
    #tiles=‘스타일’: 지도 스타일을 지정(openstreetmap,stamenterrain,cartodbpositron)

parkingMap2 = folium.Map(location = [37.4974, 126.9563],
                         zoom_start = 11,
                         zoom_control = True,
                         control_scale=True)
parkingMap2
```

###2)MarkerCluster 생성하고 지도에 추가하기

```python
#MarkerCluster 라이브러리 선언하기
#from folium.plugins import MarkerCluster
from folium.plugins import MarkerCluster

#MarkerCluster를 생성하고 지도에 추가하기
#변수명 = MarkerCluster().add_to(지도변수명)
mc = MarkerCluster().add_to(parkingMap2)
```

###3)Marker들을 생성하고 MarkerCluster에 추가하기

```python
#folium.Marker(속성들).add_to(MarkerCluster변수명)

for i, row in data.iterrows():   #전체 데이터 사용 예정
    주차장명 = row['주차장명']
    위도 = row['위도']
    경도 = row['경도']

    mk = folium.Marker(location = [위도, 경도],
                     tooltip = 주차장명,   #속성을 popup 으로 바꿔보기...다른점은?
                     icon =  folium.Icon(color='red', icon ='car', prefix='fa' ))
    mk.add_to(mc)

parkingMap2
```
